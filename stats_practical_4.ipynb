```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'minority': np.random.choice(['Yes', 'No'], 50, p=[0.3, 0.7]),
    'tenure': np.random.choice(['Yes', 'No'], 50, p=[0.5, 0.5]),
    'age': np.random.randint(30, 65, 50),
    'gender': np.random.choice(['Male', 'Female'], 50),
    'evaluation': np.round(np.random.normal(4.0, 0.5, 50), 2)
}

df = pd.DataFrame(data)
df
```

```python
'''
Calculate the percentage of visible miniorities are tenure professors
Will you say that tensure status differed if teacher was a visible minority?
'''
tenured_minority = df[(df['tenure'] == 'Yes') & (df['minority'] == 'Yes')]
minority_total = df[df['minority'] == 'Yes']

percentage = (len(tenured_minority) / len(minority_total)) * 100
print(f"Percentage of visible minorities who are tenured professors: {percentage:.2f}%")

if percentage < 50:
    print("Tenure status differs — visible minorities are less likely to be tenured.")
else:
    print("Tenure status does not differ much between visible minorities and others.")
```

```python
'''
does avg age differ by tenure? process the mean and sd for both tenured
and untenured professors
'''
tenured = df[df['tenure'] == 'Yes']['age']
untenured = df[df['tenure'] == 'No']['age']

print("Average age by tenure:")
print(f"Tenured - Mean: {tenured.mean():.2f}, Std Dev: {tenured.std():.2f}")
print(f"Untenured - Mean: {untenured.mean():.2f}, Std Dev: {untenured.std():.2f}")

# boxplot
plt.boxplot([tenured, untenured], tick_labels=['Tenured', 'Untenured'])
plt.title("Age Distribution by Tenure Status")
plt.ylabel("Age")
plt.show()
```

```python
# which graph works better and plot that graph for age variable
plt.hist(df['age'], edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
print("Age variable graph — Histogram works best for continuous numeric data")
```

```python
# define differnce b/w pyplot.bar and pyplot.barh also plot for gender variable

gender_counts = df['gender'].value_counts()

plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Distribution (Vertical Bar Chart)")
plt.ylabel("Count")
plt.show()

plt.barh(gender_counts.index, gender_counts.values)
plt.title("Gender Distribution (Horizontal Bar Chart)")
plt.xlabel("Count")
plt.show()
```

**bar**
* Creates a vertical bar chart.

* Categories are shown on the x-axis, and values (heights) on the y-axis.

* Commonly used when you have short category names or want to show growth trends.

**barh**

* Creates a horizontal bar chart.

* Categories are shown on the y-axis, and values (lengths) on the x-axis.

* Better when category names are long or you want to emphasize comparisons rather than progression.




```python
# What is the median evaluation for tenured professors?
median_eval = df[df['tenure'] == 'Yes']['evaluation'].median()
print(f"Median evaluation score for tenured professors: {median_eval:.2f}")
```
